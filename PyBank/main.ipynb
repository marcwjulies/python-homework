```python
# Marc Julies
# Import the pathlib and csv library
from pathlib import Path
import csv
# Initialize line_num variable
line_num = 0

# Set the file path
csvpath = Path('budget_data.csv')

# Initialise data structures
# Initialize list to store profit and loss data
pl_list = [] 

# Initialize metrics variables
average_sum = 0
average_sum_counter = 0;
average_change = 0
max_change = 0
min_change = 0
total_months = 0;
total = 0

# Open the input path as a file object
with open(csvpath, 'r') as csvfile:
        
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')

    # Go to the next row from the start of the file
    # (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    line_num += 1

    for row in csvreader:
        # Initialise profit and loss dictionary
        pl = {}
        pl['period'] = row[0]
        pl['profit_loss'] = row[1]
        pl['change']=0
        pl_list.append(pl)
    
    # Set total number of file entries
    total_months = len(pl_list)       
        
    # Loop through the profit and lost list
    for i in range(total_months-1):
        item = []
        nextitem = []
        item = pl_list[i]
        nextitem = pl_list[i+1]
        nextitem['change'] = int(nextitem['profit_loss']) - int(item['profit_loss'])
    
    # Calculate Metrics
   
    for i in range(total_months):
        item = []
        item = pl_list[i]
        total += int(item['profit_loss'])
        
        #Total Average Calculation
        if 'change' in item.keys():
            average_sum_counter += 1
            average_sum += item['change']
            
            #Max Change Calculation
            current_max = int(item['change'])
            if current_max > max_change:
                max_change = int(item['change'])
                max_change_str = f"Greatest Increase in Profits: {item['period']} (${item['change']})"
                
            #Min Change Calculation
            current_min = int(item['change'])
            if current_min < min_change:
                min_change = int(item['change'])
                min_change_str = f"Greatest Decrease in Profits: {item['period']} (${item['change']})"

#Average Change Calculation
average_chg = round( average_sum / average_sum_counter, 2)

#Print Results
print("Financial Analysis")
print("-------------------------")
print(f"Total Months: {total_months}")
print(f"Total: ${total}")
print(f"Average Change: ${average_chg}")
print(f"{max_change_str}")
print(f"{min_change_str}")
```
